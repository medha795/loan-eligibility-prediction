```python
# Required Libraries
import pandas as pd
import numpy as np

# Set file paths
accepted_path = "C:/Users/Medha/Projects/loan-eligibility-prediction/data/raw/accepted_2007_to_2018Q4.csv"
rejected_path = "C:/Users/Medha/Projects/loan-eligibility-prediction/data/raw/rejected_2007_to_2018Q4.csv"

# Load only needed columns from each dataset to avoid memory issues
accepted_df = pd.read_csv(accepted_path, low_memory=False, usecols=[
    'loan_amnt', 'term', 'int_rate', 'emp_length', 'home_ownership', 'annual_inc', 
    'purpose', 'addr_state', 'dti', 'delinq_2yrs', 'fico_range_high', 'fico_range_low',
    'inq_last_6mths', 'loan_status', 'application_type'
])

rejected_df = pd.read_csv(rejected_path, low_memory=False, usecols=[
    'Amount Requested', 'Application Date', 'Loan Title', 'Risk_Score',
    'Debt-To-Income Ratio', 'Zip Code', 'State', 'Employment Length', 'Policy Code'
])

# Clean column names
accepted_df.columns = accepted_df.columns.str.lower().str.replace(' ', '_')
rejected_df.columns = rejected_df.columns.str.lower().str.replace(' ', '_')

# Create 'loan_status' label
accepted_df['loan_status'] = 'Accepted'
rejected_df['loan_status'] = 'Rejected'

# Rename rejected columns to match where possible
rejected_df = rejected_df.rename(columns={
    'amount_requested': 'loan_amnt',
    'employment_length': 'emp_length',
    'state': 'addr_state',
    'debt-to-income_ratio': 'dti',
    'risk_score': 'fico_range_low'
})

# Add placeholder columns to rejected to match accepted
missing_cols = [col for col in accepted_df.columns if col not in rejected_df.columns]
for col in missing_cols:
    rejected_df[col] = np.nan

# Reorder to match accepted
rejected_df = rejected_df[accepted_df.columns]

# Combine datasets
combined_df = pd.concat([accepted_df, rejected_df], ignore_index=True)
```
